```python
import os
import json
import pandas as pd
import mysql.connector 
```

```python
# Get path of Aggregated transtion 

Aggre_path=r"C:\\Users\\USER\\Documents\\Phonepe Pulse\\pulse\\data\\aggregated\\transaction\\country\\india\\state\\"
Aggre_tran_state_list=os.listdir(Aggre_path)

# To Collect Aggregate transtion data 
clm_1= {
    "States":[],
    "Year":[],
    "Quarter":[],
    "Transation_type":[],
    "Transation_count":[],
    "Transation_amount":[]
}

for state_t in Aggre_tran_state_list:
    path_state_t=Aggre_path + state_t + "\\"
    Aggre_year_list= os.listdir(path_state_t)
    for year_t in Aggre_year_list:
        path_year_t= path_state_t +year_t+ "\\"
        Aggre_file_list=os.listdir(path_year_t)
        for file_t in Aggre_file_list:
            path_file_t= path_year_t + file_t
            collected_tran_data=open(path_file_t,"r") 
            Data= json.load(collected_tran_data)
            for x in Data['data']['transactionData']:
                Name_t=x['name']
                count_t=x['paymentInstruments'][0]['count']
                Amount_t=x['paymentInstruments'][0]['amount']
                clm_1["Transation_type"].append(Name_t)
                clm_1["Transation_count"].append(count_t)
                clm_1["Transation_amount"].append(Amount_t)
                clm_1["States"].append(state_t)
                clm_1["Year"].append(year_t)
                clm_1["Quarter"].append(int(file_t.strip(".json")))

df_Aggre_transation=pd.DataFrame(clm_1)

df_Aggre_transation["States"]=df_Aggre_transation["States"].str.replace('andaman-&-nicobar-islands','Andaman & Nicobar')
df_Aggre_transation["States"]=df_Aggre_transation["States"].str.replace("-"," ")
df_Aggre_transation["States"]=df_Aggre_transation["States"].str.title()
df_Aggre_transation["States"]=df_Aggre_transation["States"].str.replace('Dadra & Nagar Haveli & Daman & Diu','Dadra and Nagar Haveli and Daman and Diu')
```

```python
# Get path of Aggregated user 
user_path="C:\\Users\\USER\\Documents\\Phonepe Pulse\\pulse\\data\\aggregated\\user\\country\\india\\state\\"
Aggre_user_state_list=os.listdir(user_path)

# Get Aggregate user data  
clm_2= {
    "States":[],
    "Year":[],
    "Quarter":[],
    "Brand_name":[],
    "user_count":[],
    "Percentage":[]
}
for state_u in Aggre_user_state_list:
    path_state_u=user_path+state_u+"\\"
    user_year_list=os.listdir(path_state_u)
    for year_u in user_year_list:
        path_year_u=path_state_u+year_u+"\\"
        user_file_list=os.listdir(path_year_u)
        for file_u in user_file_list:
            path_file_u=path_year_u+file_u
            collected_user_data=open(path_file_u,"r")  
            user_data=json.load(collected_user_data)
            try:
                for a in user_data.get("data",{}).get("usersByDevice",[]):
                    brand_name=a.get("brand")
                    count_u=a.get("count")
                    percentage=a.get("percentage")
                    clm_2["Brand_name"].append(brand_name)
                    clm_2["user_count"].append(count_u)
                    clm_2["Percentage"].append(percentage)
                    clm_2["States"].append(state_u )
                    clm_2["Year"].append(year_u)
                    clm_2["Quarter"].append(file_u.strip(".json"))
            except KeyError as e:
                print(f"KeyError: {e}")
            except:
                pass

df_Aggre_user=pd.DataFrame(clm_2)
            
df_Aggre_user["States"]=df_Aggre_user["States"].str.replace('andaman-&-nicobar-islands','Andaman & Nicobar')
df_Aggre_user["States"]=df_Aggre_user["States"].str.replace("-"," ")
df_Aggre_user["States"]=df_Aggre_user["States"].str.title()
df_Aggre_user["States"]=df_Aggre_user["States"].str.replace('Dadra & Nagar Haveli & Daman & Diu','Dadra and Nagar Haveli and Daman and Diu')
```
